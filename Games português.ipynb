### Preenchendo Valores Ausentes

* Vizualizando o total de valores ausentes

```python
df.isna().sum()
```

* Substitui os dois nomes desconhecidos por 'unknown', acredito ser o melhor método já que não são números para determinar a média e mediana por exemplo, nome é algo único deve continuar assim. Além de ser apenas dois nomes faltantes.

```python

df['name'] = df['name'].fillna('unknown')

df['name'].isna().sum()
```

```python
df[df['yor'].isna()]
```

* Jogos sem yor tendem a ser antigos e sem muita repercusão, salvo algumas exceções.

* Por o ano de lançamento(yor) ser algo numérico acredito que se eu substituir com a mediana do ano de lancamento dos jogos de mesmo gênero e de mesma plataforma será a melhor opção.

```python
dfyor= df.pivot_table(index='genre', columns='platform' ,values='yor', aggfunc='median')
dfyor
```

```python
def aux_fill_NaN(platform,genre):
    try:
        return dfyor[platform][genre]   
    except:
        return 'erro'
```

```python
aux_fill_NaN('3DS','Puzzle')
```

```python
df['mediana_yor'] = df.apply(lambda row: aux_fill_NaN(row['platform'],row['genre']), axis=1)
df['yor']=df['yor'].fillna(df['mediana_yor'])
```

```python
df[df['yor']=='erro']
```

```python
df['yor'].isna().sum()
```

* Gênero também é algo único, não se pode determinar a média de gêneros, mesmo se substituído pelo mais frequente seria mesmo assim algo incerto, por ser apenas dois também, não acredito que seja algo que influencie muito mas mesmo assim deixarei por unknown.

```python
df['genre'] = df['genre'].fillna('unknown')

df['genre'].isna().sum()
```

```python
df[df['genre']=='unknown']
```

* Pela coincidência os valores ausentes de genre e name são os mesmos, aparenta ser apenas um jogo antigo sem muita informação, assim como os valores de yor anteriormente ausentes tendem a ser jogos antigos.


* Pensei muito sobre como lidar com valores ausentes nas colunas de críticas, mas acredito que a melhor solução seja substituir pela média já que isso não irá afetar futuras conclusões. 

```python
df['critic_score'] = df['critic_score'].fillna(df['critic_score'].mean())


display(df['critic_score'].isna().sum())
```

* Como a coluna de rating é feita por strings, e única, vou apenas substituir por 'unknown'

```python
df['rating'] = df['rating'].fillna('unknown')
df['rating'].isna().sum()
```

### Conversão de Dados

* Pela minha análise o tipo de dado da coluna yor(year of realease) deve ser modificado, pois ano é um valor inteiro. E o tipo de dado da coluna user_score deve ser um float, pois é um valor decimal, mas antes irei criar uma nova coluna tbd com valor booleano de 0 e 1 para poder lidar com a coluna original depois.

```python
#Mudança no yor
df['yor'] = df['yor'].astype(int)
df['yor'].dtype
```

```python
#mudança no user_score
df['tbd'] = 0
df['tbd'] = df.apply(lambda row: row['tbd'] + 1 if row['user_score'] == 'tbd' else row['tbd'], axis=1)
display(df['tbd'].value_counts())
df[df['user_score']=='tbd']
```

```python
#mudança no user_score
df['user_score'] = df['user_score'].replace('tbd',np.nan)
df['user_score'] = df['user_score'].astype(float)
df['user_score'].dtype
```

```python
df['user_score'] = df['user_score'].fillna(df['user_score'].mean())
df['user_score'].isna().sum()
```

```python
df.info()
```

```python
#Preparando o df para testar
dfxone=df[df['platform']=='XOne']
display(dfxone)
dfpc=df[df['platform']=='PC']
dfpc
```

* — As classificações médias dos usuários das plataformas Xbox One e PC são as mesmas.
* h0= as medias sao iguais(concorda com a tese)
* h1 as medias sao diferentes(discorda com a tese)

```python
#— As classificações médias dos usuários das plataformas Xbox One e PC são as mesmas.
#h0= as medias sao iguais(concorda com a tese)
#h1 as medias sao diferentes(discorda com a tese)
Xone= dfxone['user_score']
Pc=dfpc['user_score']
varianciaxone= np.var(Xone)
varianciapc = np.var(Pc)
display(varianciaxone,varianciapc)
from scipy.stats import levene
display(levene(Xone,Pc,center='mean'))

alpha=0.05
results= st.ttest_ind(Xone, Pc, equal_var = True)

print('p-value:', results.pvalue)

if results.pvalue < alpha:
    print("Nós rejeitamos a hipótese nula")
else:
    print("Nós não podemos rejeitar a hipótese nula")
```

* Neste teste utilizei a Hipótese sobre a Igualdade das Médias de Duas Populações, pois precisei testar as médias de duas populações estatísticas diferentes, porém deixei o equal_var=True pois a variância é próxima de acordo com o teste do levene

```python
#Preparando o df para testar
dfaction=df[df['genre']=='Action']
display(dfaction)
dfsports=df[df['genre']=='Sports']
dfsports
```

* — As classificações médias de usuários para os gêneros Action (ação) e Sports (esportes) são diferentes.

```python
#— As classificações médias de usuários para os gêneros Action (ação) e Sports (esportes) são diferentes.
#h0=as medias sao diferentes(concorda com a tese)
#h1=as medias sao iguais(discorda com a tese)
Action = dfaction['user_score']
Sports = dfsports['user_score']
vara=np.var(Action)
varb=np.var(Sports)
print(vara,varb)
from scipy.stats import levene
display(levene(Action,Sports,center='mean'))

alpha=0.05
results= st.ttest_ind(Action, Sports,equal_var = True)

print('p-value:', results.pvalue)

if results.pvalue < alpha:
    print("Nós rejeitamos a hipótese nula")
else:
    print("Nós não podemos rejeitar a hipótese nula")
```
